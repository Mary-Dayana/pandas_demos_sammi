```python
# 7. Read the total profit of all months and show it using the Bar plot. 
# The generated Bar plot must include the following properties: 
# X label name = Month number.
# Y label name = Total profit.
# Title = Company profit per month.

pyplot.bar(df['month_number'], df["total_profit"],color='blue')
plt.pyplot.xlabel("Month number")
plt.pyplot.ylabel("Total Profit")
plt.pyplot.yticks([100000, 200000, 300000, 400000, 500000])
plt.pyplot.title("Company Profit Per Month")
plt.pyplot.grid(True, linewidth= 2, linestyle="--", color= 'grey')
```

```python
# 8. Read the total profit of all of the months, and show the line plot with the following style properties:
# The generated line plot must include the following Style properties: 
# The line style should be dotted and the line color should be red.
# Show the legend at the lower-right location.
# X label name = Month number
# Y label name = Total profit.
# Title = Company profit per month.
# Add a circle marker; the line marker color = red. 
# The line width should be 3.

profit_list = df['total_profit'].tolist()
month_list = df['month_number'].tolist()
plt.pyplot.plot(month_list,profit_list,color='red',label='Profit data of last year',linestyle='dotted',marker='o',lw=3)
plt.pyplot.title('Company profit per month')
plt.pyplot.ylabel('Total Profit')
plt.pyplot.xticks(month_list)
plt.pyplot.yticks([100000, 200000, 300000, 400000, 500000])
plt.pyplot.xlabel('Month Number')
plt.pyplot.legend(loc='lower right')
```

```python
# 9. Print all of the product sales data and show it using a multi-line plot.
# Display the number of units sold per month for each product using multi-line plots. (i.e., show a separate Plotline for each product).
# The graph should look like the screenshot below:
# GRAPH BELOW SHOULD NOT HAVE TOOTHPASTE MULTIPLE TIMES IN LEGEND. DISREGARD THAT IN THE LEGEND, ONE LINE FOR EACH PRODUCT

month_list = df['month_number'].tolist()
faceCream_list=df['facecream'].tolist()
faceWash_list = df['facewash'].tolist()
toothPaste_list = df['toothpaste'].tolist()
bathSoap_list=df['bathingsoap'].tolist()
shampoo_list = df['shampoo'].tolist()
moisturizer_list=df['moisturizer'].tolist()
plt.pyplot.plot(month_list, faceCream_list, label='Face Cream Sales Data',marker='o',linewidth=3)
plt.pyplot.plot(month_list,faceWash_list, label='Face Wash Sales Data',marker='o',linewidth=3)
plt.pyplot.plot(month_list,toothPaste_list, label='ToothPaste Sales Data',marker='o',linewidth=3)
plt.pyplot.plot(month_list,bathSoap_list, label='Bath Soap Sales Data',marker='o',linewidth=3)
plt.pyplot.plot(month_list,shampoo_list, label='Shampoo Sales Data',marker='o',linewidth=3)
plt.pyplot.plot(month_list,moisturizer_list, label='Moisturizer Sales Data',marker='o',linewidth=3)
plt.pyplot.xlabel("Month Number")
plt.pyplot.ylabel("Sales Units In Numbers")
plt.pyplot.legend(loc="upper left")
plt.pyplot.xticks(month_list)
plt.pyplot.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.pyplot.title("Sales Data")
#plt.pyplot.show()
```

```python
# 10. Read “bathingsoap” sales data for each month and show it using a scatter plot.
# Add a grid to the plot. The grid line style should be “–“.

bathsoap_list = df ['bathingsoap'].tolist()
plt.pyplot.scatter(month_list, bathsoap_list, label = 'Bathingsoap Sales data', color='blue')
plt.pyplot.xlabel('Month Number')
plt.pyplot.ylabel('Number of units Sold')
plt.pyplot.legend(loc='upper left')
plt.pyplot.title('Bathingsoap Sales data')
plt.pyplot.xticks(month_list)
plt.pyplot.yticks([6000, 8000, 10000, 12000, 14000])
plt.pyplot.grid(True, linewidth= 1, linestyle="--", color= 'grey')
plt.pyplot.show()
```
